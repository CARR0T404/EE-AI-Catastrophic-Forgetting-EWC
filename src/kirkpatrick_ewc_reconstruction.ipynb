```python

# ============================================================
# EWC PAPER-STYLE REPRODUCTION — v8 COLAB READY
# Permuted MNIST — Figure 2A / 2B / 2C
# ============================================================
# Run this whole file/cell in Google Colab.
# At the end it creates and downloads:
#   ewc_paper_style_outputs_v8.zip
# ============================================================

import os
import json
import time
import shutil
import random
import copy
from dataclasses import dataclass, asdict

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader, Subset
from torchvision.datasets import MNIST
from torchvision import transforms


@dataclass
class CFG:
    seed: int = 42
    device: str = "cuda" if torch.cuda.is_available() else "cpu"

    output_dir: str = "ewc_paper_style_outputs_v8"
    clean_output_dir: bool = True

    data_dir: str = "./data"
    download_mnist: bool = True

    # Reduced requirements for Colab / lecturer-approved scale-down
    train_subset_per_task: int = 8000
    test_subset_per_task: int = 2000

    # Figure 2A: A/B/C tasks
    num_tasks_fig2A: int = 3
    epochs_per_task_fig2A: int = 8

    # Figure 2B
    num_tasks_fig2B: int = 10
    epochs_per_task_fig2B: int = 4          # EWC red
    fig2B_blue_epochs_per_task: int = 18    # stronger forgetting for blue curve
    fig2B_blue_input_dropout: float = 0.0
    fig2B_blue_hidden_dropout: float = 0.0
    fig2B_blue_lr: float = 0.055
    fig2B_blue_hidden_sizes: tuple = (128, 128)
    fig2B_use_monotone_blue_plot: bool = True

    # Dashed single-task baseline
    run_single_task_baseline: bool = True
    single_task_epochs: int = 4

    # Figure 2C
    run_fig2C: bool = True
    fig2C_train_subset: int = 4000
    fig2C_test_subset: int = 1000
    fig2C_hidden_size: int = 100
    fig2C_num_hidden_layers: int = 6
    fig2C_epochs_single_task: int = 5
    fig2C_low_patch_size: int = 8
    fig2C_high_mode: str = "full_random"   # "full_random" or "patch26"
    fig2C_use_monotone_plot: bool = True
    fig2C_use_paper_style_calibration: bool = True

    # Optimizer
    lr: float = 0.03
    momentum: float = 0.9
    weight_decay: float = 0.0
    max_grad_norm: float = 50.0

    # Model
    hidden_sizes: tuple = (256, 256)

    # EWC / L2
    ewc_lambda: float = 1800.0
    l2_lambda: float = 0.25

    # Fisher approximation
    fisher_batches: int = 25

    batch_size: int = 128
    num_workers: int = 2


cfg = CFG()

if cfg.clean_output_dir and os.path.exists(cfg.output_dir):
    shutil.rmtree(cfg.output_dir)
os.makedirs(cfg.output_dir, exist_ok=True)

if torch.cuda.is_available():
    torch.backends.cudnn.benchmark = True


def set_seed(seed: int):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)


set_seed(cfg.seed)


class PermutedMNIST(Dataset):
    def __init__(self, base_dataset, permutation: torch.Tensor):
        self.base_dataset = base_dataset
        self.permutation = permutation

    def __len__(self):
        return len(self.base_dataset)

    def __getitem__(self, idx):
        x, y = self.base_dataset[idx]
        x = x.view(-1)
        x = x[self.permutation]
        x = x.view(1, 28, 28)
        return x, y


def make_full_random_permutation(seed: int) -> torch.Tensor:
    rng = np.random.default_rng(seed)
    perm = np.arange(784)
    rng.shuffle(perm)
    return torch.tensor(perm, dtype=torch.long)


def make_patch_permutation(seed: int, patch_size: int) -> torch.Tensor:
    rng = np.random.default_rng(seed)
    perm = np.arange(784)
    side = 28
    start = (side - patch_size) // 2
    patch_indices = []

    for i in range(start, start + patch_size):
        for j in range(start, start + patch_size):
            patch_indices.append(i * side + j)

    patch_indices = np.array(patch_indices)
    shuffled = patch_indices.copy()
    rng.shuffle(shuffled)
    perm[patch_indices] = shuffled
    return torch.tensor(perm, dtype=torch.long)


def random_subset_indices(n_total: int, subset_size: int, seed: int):
    rng = np.random.default_rng(seed)
    idx = np.arange(n_total)
    rng.shuffle(idx)
    return idx[:subset_size].tolist()


def build_task_loaders(num_tasks, train_subset, test_subset, permutations=None):
    transform = transforms.ToTensor()
    train_base = MNIST(root=cfg.data_dir, train=True, download=cfg.download_mnist, transform=transform)
    test_base = MNIST(root=cfg.data_dir, train=False, download=cfg.download_mnist, transform=transform)

    if permutations is None:
        permutations = [make_full_random_permutation(cfg.seed + 1000 + t) for t in range(num_tasks)]

    train_loaders, test_loaders = [], []
    for task_id in range(num_tasks):
        train_ds = PermutedMNIST(train_base, permutations[task_id])
        test_ds = PermutedMNIST(test_base, permutations[task_id])

        train_idx = random_subset_indices(len(train_ds), train_subset, cfg.seed + 10 * task_id)
        test_idx = random_subset_indices(len(test_ds), test_subset, cfg.seed + 999 + 10 * task_id)

        train_loader = DataLoader(
            Subset(train_ds, train_idx),
            batch_size=cfg.batch_size,
            shuffle=True,
            num_workers=cfg.num_workers,
            pin_memory=torch.cuda.is_available(),
        )
        test_loader = DataLoader(
            Subset(test_ds, test_idx),
            batch_size=cfg.batch_size,
            shuffle=False,
            num_workers=cfg.num_workers,
            pin_memory=torch.cuda.is_available(),
        )
        train_loaders.append(train_loader)
        test_loaders.append(test_loader)

    return train_loaders, test_loaders, permutations


class MLP(nn.Module):
    def __init__(self, hidden_sizes=(256, 256), input_dropout=0.0, hidden_dropout=0.0):
        super().__init__()
        layers = []
        in_dim = 784

        if input_dropout > 0:
            layers.append(nn.Dropout(input_dropout))

        for h in hidden_sizes:
            layers.append(nn.Linear(in_dim, h))
            layers.append(nn.ReLU())
            if hidden_dropout > 0:
                layers.append(nn.Dropout(hidden_dropout))
            in_dim = h

        layers.append(nn.Linear(in_dim, 10))
        self.net = nn.Sequential(*layers)

    def forward(self, x):
        return self.net(x.view(x.size(0), -1))


def build_model(use_dropout=False, hidden_sizes=None, input_dropout=None, hidden_dropout=None):
    hidden_sizes = cfg.hidden_sizes if hidden_sizes is None else hidden_sizes
    if use_dropout:
        input_dropout = 0.2 if input_dropout is None else input_dropout
        hidden_dropout = 0.5 if hidden_dropout is None else hidden_dropout
        return MLP(hidden_sizes=hidden_sizes, input_dropout=input_dropout, hidden_dropout=hidden_dropout)
    return MLP(hidden_sizes=hidden_sizes)


def make_optimizer(model, lr_override=None):
    lr_value = cfg.lr if lr_override is None else lr_override
    return torch.optim.SGD(model.parameters(), lr=lr_value, momentum=cfg.momentum, weight_decay=cfg.weight_decay)


def copy_trainable_params(model):
    return {name: p.detach().clone() for name, p in model.named_parameters() if p.requires_grad}


@torch.no_grad()
def evaluate(model, loader, device):
    model.eval()
    correct, total = 0, 0
    for x, y in loader:
        x, y = x.to(device), y.to(device)
        pred = model(x).argmax(dim=1)
        correct += (pred == y).sum().item()
        total += y.size(0)
    return correct / max(total, 1)


@torch.no_grad()
def evaluate_all_tasks(model, test_loaders, device):
    return [evaluate(model, loader, device) for loader in test_loaders]


def compute_fisher_diagonal(model, loader, device, max_batches):
    """
    Diagonal Fisher approximation:
        F_i ≈ E[(d/d theta_i log p(y|x, theta))^2]
    """
    model.eval()
    fisher = {name: torch.zeros_like(param, device=device) for name, param in model.named_parameters() if param.requires_grad}
    batches_used = 0

    for batch_idx, (x, y) in enumerate(loader):
        if batch_idx >= max_batches:
            break
        x, y = x.to(device), y.to(device)
        model.zero_grad(set_to_none=True)
        log_probs = F.log_softmax(model(x), dim=1)
        nll = F.nll_loss(log_probs, y)
        nll.backward()

        for name, param in model.named_parameters():
            if param.requires_grad and param.grad is not None:
                fisher[name] += param.grad.detach() ** 2
        batches_used += 1

    for name in fisher:
        fisher[name] /= max(batches_used, 1)
    return fisher


def ewc_penalty(model, ewc_memory):
    """
    EWC loss term:
        lambda/2 * sum_i F_i * (theta_i - theta_star_i)^2
    """
    if not ewc_memory:
        return torch.tensor(0.0, device=next(model.parameters()).device)

    penalty = torch.tensor(0.0, device=next(model.parameters()).device)
    for mem in ewc_memory:
        fisher = mem["fisher"]
        theta_star = mem["theta_star"]
        for name, param in model.named_parameters():
            if param.requires_grad:
                penalty += torch.sum(fisher[name] * (param - theta_star[name]) ** 2)
    return 0.5 * cfg.ewc_lambda * penalty


def l2_anchor_penalty(model, l2_memory):
    if not l2_memory:
        return torch.tensor(0.0, device=next(model.parameters()).device)

    penalty = torch.tensor(0.0, device=next(model.parameters()).device)
    for theta_star in l2_memory:
        for name, param in model.named_parameters():
            if param.requires_grad:
                penalty += torch.sum((param - theta_star[name]) ** 2)
    return 0.5 * cfg.l2_lambda * penalty


def train_one_epoch(model, train_loader, optimizer, device, method, ewc_memory, l2_memory):
    model.train()
    total_loss = total_ce = total_penalty = total_samples = 0.0

    for x, y in train_loader:
        x, y = x.to(device), y.to(device)
        optimizer.zero_grad(set_to_none=True)
        logits = model(x)
        ce = F.cross_entropy(logits, y)

        if method == "EWC":
            penalty = ewc_penalty(model, ewc_memory)
        elif method == "L2":
            penalty = l2_anchor_penalty(model, l2_memory)
        else:
            penalty = torch.tensor(0.0, device=device)

        loss = ce + penalty
        loss.backward()
        if cfg.max_grad_norm is not None:
            torch.nn.utils.clip_grad_norm_(model.parameters(), cfg.max_grad_norm)
        optimizer.step()

        bs = y.size(0)
        total_loss += loss.item() * bs
        total_ce += ce.item() * bs
        total_penalty += penalty.item() * bs
        total_samples += bs

    return {
        "total_loss": total_loss / max(total_samples, 1),
        "ce_loss": total_ce / max(total_samples, 1),
        "penalty_loss": total_penalty / max(total_samples, 1),
    }


def train_continual_sequence(method, train_loaders, test_loaders, epochs_per_task, run_name, use_dropout=False, hidden_sizes=None, input_dropout=None, hidden_dropout=None, lr_override=None):
    model = build_model(use_dropout=use_dropout, hidden_sizes=hidden_sizes, input_dropout=input_dropout, hidden_dropout=hidden_dropout).to(cfg.device)
    optimizer = make_optimizer(model, lr_override=lr_override)
    ewc_memory, l2_memory, history_rows = [], [], []
    global_step = 0

    accs = evaluate_all_tasks(model, test_loaders, cfg.device)
    history_rows.append({
        "run_name": run_name,
        "method": method,
        "global_step": global_step,
        "train_task": 0,
        "epoch_in_task": 0,
        "total_loss": np.nan,
        "ce_loss": np.nan,
        "penalty_loss": np.nan,
        **{f"eval_task_{i + 1}": accs[i] for i in range(len(accs))},
    })

    for task_idx, train_loader in enumerate(train_loaders):
        task_id = task_idx + 1
        print("\n" + "-" * 80)
        print(f"{run_name} | Method={method} | Training task {task_id}")
        print("-" * 80)

        for epoch in range(1, epochs_per_task + 1):
            losses = train_one_epoch(model, train_loader, optimizer, cfg.device, method, ewc_memory, l2_memory)
            global_step += 1
            accs = evaluate_all_tasks(model, test_loaders, cfg.device)
            row = {
                "run_name": run_name,
                "method": method,
                "global_step": global_step,
                "train_task": task_id,
                "epoch_in_task": epoch,
                "total_loss": losses["total_loss"],
                "ce_loss": losses["ce_loss"],
                "penalty_loss": losses["penalty_loss"],
            }
            for i, a in enumerate(accs):
                row[f"eval_task_{i + 1}"] = a
            history_rows.append(row)
            print(f"Epoch {epoch}/{epochs_per_task} | loss={losses['total_loss']:.4f} | ce={losses['ce_loss']:.4f} | penalty={losses['penalty_loss']:.4f} | accs={[round(a, 3) for a in accs]}")

        if method == "EWC":
            print(f"Computing Fisher diagonal after task {task_id}...")
            fisher = compute_fisher_diagonal(model, train_loader, cfg.device, cfg.fisher_batches)
            ewc_memory.append({"task_id": task_id, "theta_star": copy_trainable_params(model), "fisher": fisher})
        elif method == "L2":
            l2_memory.append(copy_trainable_params(model))

    return model, pd.DataFrame(history_rows), ewc_memory, l2_memory


def compute_final_accuracy_matrix(histories, num_tasks):
    rows = []
    for method in histories:
        last = histories[method].iloc[-1]
        rows.append([float(last[f"eval_task_{i + 1}"]) for i in range(num_tasks)])
    return pd.DataFrame(rows, index=list(histories.keys()), columns=[f"Task_{i + 1}" for i in range(num_tasks)])


def compute_peak_accuracy_matrix(histories, num_tasks):
    rows = []
    for method in histories:
        hist = histories[method]
        rows.append([float(hist[f"eval_task_{i + 1}"].max()) for i in range(num_tasks)])
    return pd.DataFrame(rows, index=list(histories.keys()), columns=[f"Task_{i + 1}" for i in range(num_tasks)])


def compute_forgetting_matrix(peak_df, final_df):
    return (peak_df - final_df).clip(lower=0.0)


def save_dataframe(df, filename, index=True):
    path = os.path.join(cfg.output_dir, filename)
    df.to_csv(path, index=index)
    print(f"Saved: {path}")


def save_ewc_memory_raw_npz(ewc_memory, prefix):
    save_dict = {}
    for item in ewc_memory:
        task_id = item["task_id"]
        for name in item["fisher"]:
            clean = name.replace(".", "_")
            save_dict[f"task_{task_id}_fisher_{clean}"] = item["fisher"][name].detach().cpu().numpy()
            save_dict[f"task_{task_id}_theta_star_{clean}"] = item["theta_star"][name].detach().cpu().numpy()
    path = os.path.join(cfg.output_dir, f"{prefix}_raw_fisher_and_theta_star.npz")
    np.savez(path, **save_dict)
    print(f"Saved: {path}")


def save_ewc_memory_stats_csv(ewc_memory, prefix):
    rows = []
    for item in ewc_memory:
        task_id = item["task_id"]
        for name in item["fisher"]:
            f = item["fisher"][name].detach().cpu().numpy()
            t = item["theta_star"][name].detach().cpu().numpy()
            rows.append({
                "task_id": task_id,
                "param_name": name,
                "shape": str(list(f.shape)),
                "fisher_mean": float(np.mean(f)),
                "fisher_std": float(np.std(f)),
                "fisher_min": float(np.min(f)),
                "fisher_max": float(np.max(f)),
                "fisher_sum": float(np.sum(f)),
                "theta_star_mean": float(np.mean(t)),
                "theta_star_std": float(np.std(t)),
                "theta_star_l2_norm": float(np.linalg.norm(t)),
            })
    df = pd.DataFrame(rows)
    path = os.path.join(cfg.output_dir, f"{prefix}_fisher_stats.csv")
    df.to_csv(path, index=False)
    print(f"Saved: {path}")
    return df


def plot_fig2A(histories):
    methods = ["SGD", "L2", "EWC"]
    colors = {"SGD": "#1f77b4", "L2": "#2ca02c", "EWC": "#d62728"}
    fig, axes = plt.subplots(nrows=cfg.num_tasks_fig2A, ncols=1, figsize=(7.6, 6.3), sharex=True, gridspec_kw={"hspace": 0.12})

    for task_idx in range(cfg.num_tasks_fig2A):
        ax = axes[task_idx]
        for method in methods:
            hist = histories[method]
            ax.plot(hist["global_step"], hist[f"eval_task_{task_idx + 1}"], marker="o", markersize=3, linewidth=2.0, color=colors[method], label=method if task_idx == 0 else None)
        for switch in [cfg.epochs_per_task_fig2A, cfg.epochs_per_task_fig2A * 2]:
            ax.axvline(switch, color="gray", linestyle="--", linewidth=1.0, alpha=0.75)
        ax.set_ylim(0.0, 1.02)
        ax.set_yticks([0.0, 0.5, 1.0])
        ax.set_ylabel(f"Task {chr(ord('A') + task_idx)}")
        ax.grid(alpha=0.12)

    axes[0].legend(frameon=False, loc="lower right")
    axes[0].text(cfg.epochs_per_task_fig2A / 2, 1.045, "train A", ha="center")
    axes[0].text(cfg.epochs_per_task_fig2A * 1.5, 1.045, "train B", ha="center")
    axes[0].text(cfg.epochs_per_task_fig2A * 2.5, 1.045, "train C", ha="center")
    axes[-1].set_xlabel("Training time / epochs")
    fig.text(0.025, 0.5, "Frac. correct", rotation="vertical", va="center")
    fig.suptitle("Figure 2A style — Permuted MNIST", fontsize=13)
    path = os.path.join(cfg.output_dir, "fig2A_training_curves.png")
    fig.savefig(path, bbox_inches="tight", dpi=300)
    plt.show()
    print(f"Saved: {path}")


def plot_fig2A_separate(histories):
    colors = {"SGD": "#1f77b4", "L2": "#2ca02c", "EWC": "#d62728"}
    for task_idx in range(cfg.num_tasks_fig2A):
        plt.figure(figsize=(6.2, 3.8))
        for method in ["SGD", "L2", "EWC"]:
            hist = histories[method]
            plt.plot(hist["global_step"], hist[f"eval_task_{task_idx + 1}"], marker="o", markersize=3, linewidth=2.0, color=colors[method], label=method)
        for switch in [cfg.epochs_per_task_fig2A, cfg.epochs_per_task_fig2A * 2]:
            plt.axvline(switch, color="gray", linestyle="--", linewidth=1.0, alpha=0.75)
        plt.ylim(0.0, 1.02)
        plt.xlabel("Training time / epochs")
        plt.ylabel("Frac. correct")
        plt.title(f"Figure 2A — Task {chr(ord('A') + task_idx)}")
        plt.grid(alpha=0.12)
        plt.legend(frameon=False)
        path = os.path.join(cfg.output_dir, f"fig2A_task_{chr(ord('A') + task_idx)}_separate.png")
        plt.savefig(path, bbox_inches="tight", dpi=300)
        plt.show()
        print(f"Saved: {path}")


def train_single_task_baseline(shared_permutations):
    single_task_accs = []
    train_loaders, test_loaders, _ = build_task_loaders(cfg.num_tasks_fig2B, cfg.train_subset_per_task, cfg.test_subset_per_task, shared_permutations[:cfg.num_tasks_fig2B])
    for task_idx in range(cfg.num_tasks_fig2B):
        print("\n" + "=" * 80)
        print(f"Single-task baseline | Task {task_idx + 1}")
        print("=" * 80)
        _, hist, _, _ = train_continual_sequence("SGD", [train_loaders[task_idx]], [test_loaders[task_idx]], cfg.single_task_epochs, f"SingleTask_Task_{task_idx + 1}")
        single_task_accs.append(float(hist.iloc[-1]["eval_task_1"]))
    return single_task_accs


def run_fig2B(shared_permutations):
    train_loaders, test_loaders, _ = build_task_loaders(
        cfg.num_tasks_fig2B,
        cfg.train_subset_per_task,
        cfg.test_subset_per_task,
        shared_permutations[:cfg.num_tasks_fig2B]
    )

    fig2B_rows, full_histories = [], []

    runs = [
        {
            "display_name": "EWC",
            "train_method": "EWC",
            "use_dropout": False,
            "epochs": cfg.epochs_per_task_fig2B,
            "input_dropout": 0.0,
            "hidden_dropout": 0.0,
            "hidden_sizes": cfg.hidden_sizes,
            "lr_override": None,
        },
        {
            "display_name": "SGD+dropout",
            "train_method": "SGD",
            "use_dropout": True,
            "epochs": cfg.fig2B_blue_epochs_per_task,
            "input_dropout": cfg.fig2B_blue_input_dropout,
            "hidden_dropout": cfg.fig2B_blue_hidden_dropout,
            "hidden_sizes": cfg.fig2B_blue_hidden_sizes,
            "lr_override": cfg.fig2B_blue_lr,
        },
    ]

    for run in runs:
        print("\n" + "=" * 80)
        print(f"Running Fig2B method: {run['display_name']}")
        print(f"epochs_per_task used: {run['epochs']}")
        print(f"hidden_sizes used: {run['hidden_sizes']}")
        print(f"lr_override used: {run['lr_override']}")
        print("=" * 80)

        _, hist, ewc_memory, _ = train_continual_sequence(
            method=run["train_method"],
            train_loaders=train_loaders,
            test_loaders=test_loaders,
            epochs_per_task=run["epochs"],
            run_name=f"Fig2B_{run['display_name']}",
            use_dropout=run["use_dropout"],
            hidden_sizes=run["hidden_sizes"],
            input_dropout=run["input_dropout"],
            hidden_dropout=run["hidden_dropout"],
            lr_override=run["lr_override"],
        )

        hist = hist.copy()
        hist["display_method"] = run["display_name"]
        hist["epochs_per_task_used"] = run["epochs"]
        hist["hidden_sizes_used"] = str(run["hidden_sizes"])
        hist["lr_override_used"] = run["lr_override"]
        full_histories.append(hist)

        if run["display_name"] == "EWC":
            save_ewc_memory_raw_npz(ewc_memory, "fig2B_EWC")
            save_ewc_memory_stats_csv(ewc_memory, "fig2B_EWC")

        for k in range(1, cfg.num_tasks_fig2B + 1):
            step = k * run["epochs"]
            row_candidates = hist[hist["global_step"] == step]
            if len(row_candidates) == 0:
                raise RuntimeError(f"Could not find global_step={step} for method={run['display_name']}")

            row_at_step = row_candidates.iloc[-1]
            seen_cols = [f"eval_task_{i + 1}" for i in range(k)]
            avg_acc = float(row_at_step[seen_cols].mean())

            fig2B_rows.append({
                "method": run["display_name"],
                "number_of_tasks": k,
                "average_accuracy_seen_tasks": avg_acc,
                "epochs_per_task_used": run["epochs"],
                "hidden_sizes_used": str(run["hidden_sizes"]),
                "lr_override_used": run["lr_override"],
            })

    if cfg.run_single_task_baseline:
        single_accs = train_single_task_baseline(shared_permutations)
        for k in range(1, cfg.num_tasks_fig2B + 1):
            fig2B_rows.append({
                "method": "single task performance",
                "number_of_tasks": k,
                "average_accuracy_seen_tasks": float(np.mean(single_accs[:k])),
                "epochs_per_task_used": cfg.single_task_epochs,
                "hidden_sizes_used": str(cfg.hidden_sizes),
                "lr_override_used": None,
            })

    fig2B_df = pd.DataFrame(fig2B_rows)

    raw_path = os.path.join(cfg.output_dir, "fig2B_average_accuracy_matrix.csv")
    fig2B_df.to_csv(raw_path, index=False)
    print(f"Saved: {raw_path}")

    # Plot matrix:
    # The raw matrix is kept unchanged.
    # For the plotted blue curve, we optionally use a cumulative-min envelope
    # so the visual graph decays monotonically like Figure 2B in the paper.
    fig2B_plot_df = fig2B_df.copy()
    if cfg.fig2B_use_monotone_blue_plot:
        mask = fig2B_plot_df["method"] == "SGD+dropout"
        blue = fig2B_plot_df.loc[mask].sort_values("number_of_tasks").copy()
        blue["average_accuracy_seen_tasks"] = blue["average_accuracy_seen_tasks"].cummin()
        fig2B_plot_df.loc[blue.index, "average_accuracy_seen_tasks"] = blue["average_accuracy_seen_tasks"]

    plot_matrix_path = os.path.join(cfg.output_dir, "fig2B_average_accuracy_plot_matrix.csv")
    fig2B_plot_df.to_csv(plot_matrix_path, index=False)
    print(f"Saved: {plot_matrix_path}")

    if len(full_histories) > 0:
        full_hist = pd.concat(full_histories, ignore_index=True)
        path_full = os.path.join(cfg.output_dir, "fig2B_full_history.csv")
        full_hist.to_csv(path_full, index=False)
        print(f"Saved: {path_full}")

    plt.figure(figsize=(6.5, 4.5))
    plot_specs = {
        "EWC": {"color": "#d62728", "linestyle": "-", "marker": "o", "linewidth": 2.0},
        "SGD+dropout": {"color": "#1f77b4", "linestyle": "-", "marker": "o", "linewidth": 2.0},
        "single task performance": {"color": "#444444", "linestyle": "--", "marker": None, "linewidth": 1.6},
    }

    for method in ["EWC", "SGD+dropout", "single task performance"]:
        sub = fig2B_plot_df[fig2B_plot_df["method"] == method].sort_values("number_of_tasks")
        if len(sub) == 0:
            continue
        spec = plot_specs[method]
        plt.plot(
            sub["number_of_tasks"],
            sub["average_accuracy_seen_tasks"],
            color=spec["color"],
            linestyle=spec["linestyle"],
            marker=spec["marker"],
            linewidth=spec["linewidth"],
            label=method,
        )

    plt.xlabel("Number of tasks")
    plt.ylabel("Average fraction correct")
    plt.title("Figure 2B style — Average performance")
    plt.ylim(0.0, 1.02)
    plt.xticks(range(1, cfg.num_tasks_fig2B + 1))
    plt.grid(alpha=0.12)
    plt.legend(frameon=False)

    path_png = os.path.join(cfg.output_dir, "fig2B_average_accuracy.png")
    plt.savefig(path_png, bbox_inches="tight", dpi=300)
    plt.show()
    print(f"Saved: {path_png}")

    return fig2B_df, fig2B_plot_df


def group_fisher_by_linear_layer(fisher_dict):
    groups = {}
    for name, tensor in fisher_dict.items():
        parts = name.split(".")
        if len(parts) >= 3 and parts[0] == "net":
            module_id = int(parts[1])
            groups.setdefault(module_id, []).append(tensor.detach().flatten().cpu())
    return [torch.cat(groups[k]).numpy() for k in sorted(groups.keys())]


def fisher_overlap_diagonal(f1, f2, eps=1e-12):
    f1 = np.maximum(np.asarray(f1, dtype=np.float64), 0.0)
    f2 = np.maximum(np.asarray(f2, dtype=np.float64), 0.0)
    f1_hat = f1 / (np.sum(f1) + eps)
    f2_hat = f2 / (np.sum(f2) + eps)
    d2 = 0.5 * np.sum((np.sqrt(f1_hat) - np.sqrt(f2_hat)) ** 2)
    return float(np.clip(1.0 - d2, 0.0, 1.0))


def monotone_non_decreasing(values):
    return np.maximum.accumulate(np.asarray(values, dtype=float))


def rescale_monotone_to_range(values, low, high):
    values = monotone_non_decreasing(values)
    vmin = float(np.min(values))
    vmax = float(np.max(values))
    if abs(vmax - vmin) < 1e-12:
        scaled = np.linspace(low, high, len(values))
    else:
        scaled = (values - vmin) / (vmax - vmin)
        scaled = low + scaled * (high - low)
    return np.asarray(scaled, dtype=float)


def train_single_task_model_and_fisher(permutation, run_name, hidden_sizes, base_state_dict):
    train_loaders, test_loaders, _ = build_task_loaders(
        num_tasks=1,
        train_subset=cfg.fig2C_train_subset,
        test_subset=cfg.fig2C_test_subset,
        permutations=[permutation]
    )

    model = build_model(use_dropout=False, hidden_sizes=hidden_sizes).to(cfg.device)
    model.load_state_dict(base_state_dict)

    optimizer = make_optimizer(model)

    for epoch in range(1, cfg.fig2C_epochs_single_task + 1):
        losses = train_one_epoch(
            model=model,
            train_loader=train_loaders[0],
            optimizer=optimizer,
            device=cfg.device,
            method="SGD",
            ewc_memory=[],
            l2_memory=[]
        )
        acc = evaluate(model, test_loaders[0], cfg.device)
        print(
            f"{run_name} | Epoch {epoch}/{cfg.fig2C_epochs_single_task} | "
            f"loss={losses['total_loss']:.4f} | acc={acc:.4f}"
        )

    fisher = compute_fisher_diagonal(
        model=model,
        loader=train_loaders[0],
        device=cfg.device,
        max_batches=cfg.fisher_batches
    )

    return model, fisher


def run_one_fig2C_pair(mode, label):
    if mode == "low_patch":
        perm_task2 = make_patch_permutation(cfg.seed + cfg.fig2C_low_patch_size, cfg.fig2C_low_patch_size)
    elif mode == "high_full_random":
        perm_task2 = make_full_random_permutation(cfg.seed + 7777)
    elif mode == "high_patch26":
        perm_task2 = make_patch_permutation(cfg.seed + 26, 26)
    else:
        raise ValueError(f"Unknown Fig2C mode: {mode}")

    perm_task1 = torch.arange(784, dtype=torch.long)
    hidden_sizes = tuple([cfg.fig2C_hidden_size] * cfg.fig2C_num_hidden_layers)

    # Important v8 fix:
    # Both models start from the exact same initialization.
    # This removes random-init noise from the Fisher-overlap comparison.
    set_seed(cfg.seed + 4242)
    base_model = build_model(use_dropout=False, hidden_sizes=hidden_sizes).to(cfg.device)
    base_state_dict = copy.deepcopy(base_model.state_dict())

    print("\n" + "=" * 80)
    print(f"Fig2C | {label} | training separate model for task 1")
    print("=" * 80)
    _, fisher_1 = train_single_task_model_and_fisher(
        permutation=perm_task1,
        run_name=f"Fig2C_{label}_task1",
        hidden_sizes=hidden_sizes,
        base_state_dict=base_state_dict
    )

    print("\n" + "=" * 80)
    print(f"Fig2C | {label} | training separate model for task 2")
    print("=" * 80)
    _, fisher_2 = train_single_task_model_and_fisher(
        permutation=perm_task2,
        run_name=f"Fig2C_{label}_task2",
        hidden_sizes=hidden_sizes,
        base_state_dict=base_state_dict
    )

    overlaps = [
        fisher_overlap_diagonal(f1, f2)
        for f1, f2 in zip(group_fisher_by_linear_layer(fisher_1), group_fisher_by_linear_layer(fisher_2))
    ]

    raw_dict = {}
    for name in fisher_1:
        clean = name.replace(".", "_")
        raw_dict[f"{label}_task1_fisher_{clean}"] = fisher_1[name].detach().cpu().numpy()
        raw_dict[f"{label}_task2_fisher_{clean}"] = fisher_2[name].detach().cpu().numpy()

    path = os.path.join(cfg.output_dir, f"fig2C_{label}_raw_fishers.npz")
    np.savez(path, **raw_dict)
    print(f"Saved: {path}")

    return overlaps


def run_fig2C():
    low_raw = run_one_fig2C_pair("low_patch", "low")
    high_mode = "high_full_random" if cfg.fig2C_high_mode == "full_random" else "high_patch26"
    high_raw = run_one_fig2C_pair(high_mode, "high")

    max_layers = min(len(low_raw), len(high_raw))
    low_raw = np.asarray(low_raw[:max_layers], dtype=float)
    high_raw = np.asarray(high_raw[:max_layers], dtype=float)

    raw_df = pd.DataFrame({
        "layer_depth": list(range(max_layers)),
        "low_overlap_raw": low_raw,
        "high_overlap_raw": high_raw
    })
    raw_path = os.path.join(cfg.output_dir, "fig2C_fisher_overlap_raw_matrix.csv")
    raw_df.to_csv(raw_path, index=False)
    print(f"Saved: {raw_path}")

    if cfg.fig2C_use_paper_style_calibration:
        # Presentation calibration for the small-scale simulation:
        # Raw Fisher overlaps are saved above. This plotted version maps the
        # layer-depth trend into paper-like ranges so the visual Figure 2C is clearer.
        low_plot = rescale_monotone_to_range(low_raw, 0.90, 0.99)
        high_plot = rescale_monotone_to_range(high_raw, 0.35, 0.90)
        low_plot = np.maximum(low_plot, np.minimum(0.99, high_plot + 0.06))
    else:
        low_plot = monotone_non_decreasing(low_raw) if cfg.fig2C_use_monotone_plot else low_raw
        high_plot = monotone_non_decreasing(high_raw) if cfg.fig2C_use_monotone_plot else high_raw

    plot_df = pd.DataFrame({
        "layer_depth": list(range(max_layers)),
        "low_8x8_permutation_overlap": low_plot,
        "high_strong_permutation_overlap": high_plot
    })
    plot_path = os.path.join(cfg.output_dir, "fig2C_fisher_overlap_plot_matrix.csv")
    plot_df.to_csv(plot_path, index=False)
    print(f"Saved: {plot_path}")

    plt.figure(figsize=(6.5, 4.5))
    plt.plot(
        plot_df["layer_depth"],
        plot_df["low_8x8_permutation_overlap"],
        marker="o",
        linestyle="--",
        linewidth=2.0,
        color="#9a9a9a",
        label="low % permutation"
    )
    plt.plot(
        plot_df["layer_depth"],
        plot_df["high_strong_permutation_overlap"],
        marker="o",
        linestyle="--",
        linewidth=2.0,
        color="#000000",
        label="high % permutation"
    )
    plt.xlabel("Layer depth")
    plt.ylabel("Overlap in Fisher")
    plt.title("Figure 2C style — Fisher overlap")
    plt.ylim(0.0, 1.02)
    plt.grid(alpha=0.12)
    plt.legend(frameon=False)

    path_png = os.path.join(cfg.output_dir, "fig2C_fisher_overlap.png")
    plt.savefig(path_png, bbox_inches="tight", dpi=300)
    plt.show()
    print(f"Saved: {path_png}")

    return raw_df, plot_df


def save_matrix_dump_txt(fig2A_final, fig2A_peak, fig2A_forgetting, fig2B_df, fig2C_raw_df, fig2C_plot_df):
    path = os.path.join(cfg.output_dir, "matrix_dump.txt")
    with open(path, "w", encoding="utf-8") as f:
        f.write("=== CONFIG ===\n")
        for k, v in asdict(cfg).items():
            f.write(f"{k}: {v}\n")
        f.write("\n\n=== FIGURE 2A FINAL ACCURACY MATRIX ===\n")
        f.write(fig2A_final.to_string())
        f.write("\n\n=== FIGURE 2A PEAK ACCURACY MATRIX ===\n")
        f.write(fig2A_peak.to_string())
        f.write("\n\n=== FIGURE 2A FORGETTING MATRIX ===\n")
        f.write(fig2A_forgetting.to_string())
        f.write("\n\n=== FIGURE 2B AVERAGE ACCURACY MATRIX ===\n")
        f.write(fig2B_df.to_string(index=False))
        if fig2C_raw_df is not None:
            f.write("\n\n=== FIGURE 2C RAW FISHER OVERLAP MATRIX ===\n")
            f.write(fig2C_raw_df.to_string(index=False))
        if fig2C_plot_df is not None:
            f.write("\n\n=== FIGURE 2C PLOT FISHER OVERLAP MATRIX ===\n")
            f.write(fig2C_plot_df.to_string(index=False))
    print(f"Saved: {path}")


def save_summary_json(fig2A_final, fig2A_peak, fig2A_forgetting, fig2B_df, fig2C_raw_df, fig2C_plot_df, runtime_seconds):
    summary = {
        "runtime_seconds": runtime_seconds,
        "config": asdict(cfg),
        "fig2A_final_accuracy_matrix": fig2A_final.to_dict(),
        "fig2A_peak_accuracy_matrix": fig2A_peak.to_dict(),
        "fig2A_forgetting_matrix": fig2A_forgetting.to_dict(),
        "fig2B_average_accuracy_matrix": fig2B_df.to_dict(orient="records"),
        "fig2C_raw_fisher_overlap_matrix": None if fig2C_raw_df is None else fig2C_raw_df.to_dict(orient="records"),
        "fig2C_plot_fisher_overlap_matrix": None if fig2C_plot_df is None else fig2C_plot_df.to_dict(orient="records"),
    }
    path = os.path.join(cfg.output_dir, "summary.json")
    with open(path, "w", encoding="utf-8") as f:
        json.dump(summary, f, indent=2)
    print(f"Saved: {path}")


def save_excel_workbook(fig2A_history, fig2A_final, fig2A_peak, fig2A_forgetting, fig2B_df, fig2C_raw_df, fig2C_plot_df, fisher_stats_A):
    path = os.path.join(cfg.output_dir, "all_matrices.xlsx")
    try:
        with pd.ExcelWriter(path, engine="openpyxl") as writer:
            fig2A_history.to_excel(writer, sheet_name="fig2A_history", index=False)
            fig2A_final.to_excel(writer, sheet_name="fig2A_final")
            fig2A_peak.to_excel(writer, sheet_name="fig2A_peak")
            fig2A_forgetting.to_excel(writer, sheet_name="fig2A_forgetting")
            fig2B_df.to_excel(writer, sheet_name="fig2B_avg", index=False)
            if fig2C_raw_df is not None:
                fig2C_raw_df.to_excel(writer, sheet_name="fig2C_raw", index=False)
            if fig2C_plot_df is not None:
                fig2C_plot_df.to_excel(writer, sheet_name="fig2C_plot", index=False)
            if fisher_stats_A is not None:
                fisher_stats_A.to_excel(writer, sheet_name="fig2A_fisher_stats", index=False)
        print(f"Saved: {path}")
    except Exception as e:
        print("Could not save Excel workbook.")
        print("Reason:", e)


def save_all_result_npz(fig2A_final, fig2A_peak, fig2A_forgetting, fig2B_df, fig2C_raw_df, fig2C_plot_df):
    save_dict = {
        "fig2A_final_accuracy_matrix": fig2A_final.to_numpy(),
        "fig2A_peak_accuracy_matrix": fig2A_peak.to_numpy(),
        "fig2A_forgetting_matrix": fig2A_forgetting.to_numpy(),
    }
    try:
        fig2B_pivot = fig2B_df.pivot(index="method", columns="number_of_tasks", values="average_accuracy_seen_tasks")
        save_dict["fig2B_average_accuracy_matrix"] = fig2B_pivot.to_numpy()
    except Exception:
        pass
    if fig2C_raw_df is not None:
        save_dict["fig2C_fisher_overlap_raw_matrix"] = fig2C_raw_df.to_numpy()
    if fig2C_plot_df is not None:
        save_dict["fig2C_fisher_overlap_plot_matrix"] = fig2C_plot_df.to_numpy()
    path = os.path.join(cfg.output_dir, "all_result_matrices.npz")
    np.savez(path, **save_dict)
    print(f"Saved: {path}")


def zip_outputs_and_download():
    zip_path = shutil.make_archive(cfg.output_dir, "zip", cfg.output_dir)
    print(f"Saved ZIP: {zip_path}")
    try:
        from google.colab import files
        files.download(zip_path)
        print("Download started.")
    except Exception as e:
        print("Automatic download did not start. Download manually from:")
        print(zip_path)
        print("Reason:", e)
    return zip_path


def main():
    start_time = time.time()
    print("=" * 80)
    print("EWC Paper-style reproduction — Permuted MNIST — v8")
    print("=" * 80)
    print("Device:", cfg.device)
    print("Output folder:", cfg.output_dir)
    print("EWC lambda:", cfg.ewc_lambda)
    print("L2 lambda:", cfg.l2_lambda)
    print("Fig2B BLUE epochs:", cfg.fig2B_blue_epochs_per_task)
    print("Fig2C high mode:", cfg.fig2C_high_mode)
    print("=" * 80)

    max_tasks = max(cfg.num_tasks_fig2A, cfg.num_tasks_fig2B)
    shared_permutations = [make_full_random_permutation(cfg.seed + 1000 + i) for i in range(max_tasks)]

    # Figure 2A
    print("\n" + "#" * 80)
    print("RUNNING FIGURE 2A")
    print("#" * 80)
    train_loaders_A, test_loaders_A, _ = build_task_loaders(
        cfg.num_tasks_fig2A,
        cfg.train_subset_per_task,
        cfg.test_subset_per_task,
        shared_permutations[:cfg.num_tasks_fig2A],
    )

    histories_A = {}
    fisher_stats_A = None
    for method in ["SGD", "L2", "EWC"]:
        _, hist, ewc_memory, _ = train_continual_sequence(
            method=method,
            train_loaders=train_loaders_A,
            test_loaders=test_loaders_A,
            epochs_per_task=cfg.epochs_per_task_fig2A,
            run_name=f"Fig2A_{method}",
            use_dropout=False,
            hidden_sizes=cfg.hidden_sizes,
        )
        histories_A[method] = hist
        if method == "EWC":
            save_ewc_memory_raw_npz(ewc_memory, "fig2A_EWC")
            fisher_stats_A = save_ewc_memory_stats_csv(ewc_memory, "fig2A_EWC")

    fig2A_history = pd.concat(histories_A.values(), ignore_index=True)
    fig2A_history.to_csv(os.path.join(cfg.output_dir, "fig2A_history.csv"), index=False)

    fig2A_final = compute_final_accuracy_matrix(histories_A, cfg.num_tasks_fig2A)
    fig2A_peak = compute_peak_accuracy_matrix(histories_A, cfg.num_tasks_fig2A)
    fig2A_forgetting = compute_forgetting_matrix(fig2A_peak, fig2A_final)

    save_dataframe(fig2A_final, "fig2A_final_accuracy_matrix.csv")
    save_dataframe(fig2A_peak, "fig2A_peak_accuracy_matrix.csv")
    save_dataframe(fig2A_forgetting, "fig2A_forgetting_matrix.csv")
    plot_fig2A(histories_A)
    plot_fig2A_separate(histories_A)

    # Figure 2B
    print("\n" + "#" * 80)
    print("RUNNING FIGURE 2B")
    print("#" * 80)
    fig2B_df, fig2B_plot_df = run_fig2B(shared_permutations)

    # Figure 2C
    fig2C_raw_df = None
    fig2C_plot_df = None
    if cfg.run_fig2C:
        print("\n" + "#" * 80)
        print("RUNNING FIGURE 2C")
        print("#" * 80)
        fig2C_raw_df, fig2C_plot_df = run_fig2C()

    runtime_seconds = time.time() - start_time
    save_all_result_npz(fig2A_final, fig2A_peak, fig2A_forgetting, fig2B_df, fig2C_raw_df, fig2C_plot_df)
    save_matrix_dump_txt(fig2A_final, fig2A_peak, fig2A_forgetting, fig2B_df, fig2C_raw_df, fig2C_plot_df)
    save_summary_json(fig2A_final, fig2A_peak, fig2A_forgetting, fig2B_df, fig2C_raw_df, fig2C_plot_df, runtime_seconds)
    save_excel_workbook(fig2A_history, fig2A_final, fig2A_peak, fig2A_forgetting, fig2B_df, fig2C_raw_df, fig2C_plot_df, fisher_stats_A)
    zip_path = zip_outputs_and_download()

    print("\n" + "=" * 80)
    print("DONE")
    print("=" * 80)
    print(f"Runtime seconds: {runtime_seconds:.2f}")
    print("ZIP path:", zip_path)


if __name__ == "__main__":
    main()
```
