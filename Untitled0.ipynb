```python
import gym
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
import numpy as np
import random
from collections import deque
import matplotlib.pyplot as plt # ייבוא ספריית גרפים

# --- הגדרות רשת הנוירונים (DQN) לפי ארכיטקטורת Atari מהמאמר ---
class AtariDQN(nn.Module):
    def __init__(self, action_size):
        super(AtariDQN, self).__init__()
        # שכבות קונבולוציה כפי שמופיעות בנספח המאמר [cite: 356-358]
        self.conv1 = nn.Conv2d(4, 32, kernel_size=8, stride=4)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=4, stride=2)
        self.conv3 = nn.Conv2d(64, 64, kernel_size=3, stride=1)
        # שכבה מלאה רחבה (1024 יחידות) להגברת קיבולת הרשת [cite: 358-359]
        self.fc1 = nn.Linear(7 * 7 * 64, 1024)
        self.fc2 = nn.Linear(1024, action_size)

    def forward(self, x):
        # שימוש ב-ReLU כפונקציית אקטיבציה [cite: 339]
        x = F.relu(self.conv1(x))
        x = F.relu(self.conv2(x))
        x = F.relu(self.conv3(x))
        x = x.view(x.size(0), -1)
        x = F.relu(self.fc1(x))
        return self.fc2(x)

# --- ניהול זיכרון חוויות (Replay Buffer) ---
class ReplayBuffer:
    def __init__(self, capacity):
        self.buffer = deque(maxlen=capacity)

    def push(self, state, action, reward, next_state, done):
        self.buffer.append((state, action, reward, next_state, done))

    def sample(self, batch_size):
        # דגימה רנדומלית כדי להפוך את החוויות לבלתי תלויות (Uncorrelated) [cite: 165]
        state, action, reward, next_state, done = zip(*random.sample(self.buffer, batch_size))
        return np.array(state), action, reward, np.array(next_state), done

    def __len__(self):
        return len(self.buffer)

# --- מימוש ליבת אלגוריתם ה-EWC (הקפיץ האלסטי) ---
class EWC:
    def __init__(self, model, device):
        self.model = model
        self.device = device
        # שמירת ערכי המשקולות האופטימליים של המשימה הקודמת (Anchor Point) [cite: 65, 162]
        self.params = {n: p.clone().detach() for n, p in model.named_parameters() if p.requires_grad}
        self.fisher = {}

    def compute_fisher(self, replay_buffer, num_samples=100):
        # חישוב אלכסון מטריצת פישר המייצגת את חשיבות המשקולות למשימה הקודמת [cite: 65, 161]
        self.model.eval()
        fisher = {n: torch.zeros_like(p) for n, p in self.model.named_parameters() if p.requires_grad}

        for _ in range(num_samples):
            # דגימת מיני-באצ'ים מהזיכרון לצורך חישוב הנגזרות [cite: 403]
            state, action, _, _, _ = replay_buffer.sample(32)
            state = torch.FloatTensor(state).to(self.device)

            self.model.zero_grad()
            output = self.model(state)
            log_probs = F.log_softmax(output, dim=1)
            # חישוב הנגזרת של הלוג-הסתברות (הערכת פישר) [cite: 65, 1566]
            loss = log_probs[range(32), action].mean()
            loss.backward()

            for n, p in self.model.named_parameters():
                if p.requires_grad:
                    fisher[n].data += p.grad.data ** 2 / num_samples

        self.fisher = {n: f.detach() for n, f in fisher.items()}

    def penalty(self, model):
        # חישוב ה"קפיץ": ענישה ריבועית המרוסנת על ידי מטריצת פישר [cite: 43, 68]
        loss = 0
        for n, p in model.named_parameters():
            if n in self.fisher:
                # ככל שמשקולת חשובה יותר (פישר גבוה), ה"קפיץ" מושך אותה חזק יותר לערכה המקורי [cite: 232, 237]
                loss += (self.fisher[n] * (p - self.params[n]) ** 2).sum()
        return loss

# --- לופ אימון מלא המשלב DQN ו-EWC ---
def train_continual_agent():
    # הגדרות היפר-פרמטרים מתוך טבלאות המאמר [cite: 403]
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    action_size = 18  # סט פעולות מלא של אטארי [cite: 374]
    model = AtariDQN(action_size).to(device)
    optimizer = optim.RMSprop(model.parameters(), lr=0.00025, alpha=0.95)

    # שלב 1: אימון על המשימה הראשונה (למשל המשחק Breakout)
    print("מתחיל אימון על משימה א'...")
    replay_buffer_a = ReplayBuffer(50000)
    # כאן יבוצע לופ אימון DQN סטנדרטי...
    # (לצורך הקוד נניח שהסוכן סיים להתאמן על משימה א')

    # Populate replay_buffer_a with some dummy data for demonstration
    print("ממלא באפר חוויות למשימה א' בנתוני דמה...")
    dummy_state = np.zeros((4, 84, 84)) # Example state shape for Atari (4 frames, 84x84 grayscale)
    for _ in range(100): # Add 100 dummy experiences
        action = random.randint(0, action_size - 1)
        reward = random.uniform(-1, 1)
        done = random.choice([True, False])
        replay_buffer_a.push(dummy_state, action, reward, dummy_state, done)
    print(f"באפר חוויות למשימה א' מכיל כעת {len(replay_buffer_a)} חוויות.")

    # שלב 2: הפעלת EWC - גיבוש הידע (Consolidation) [cite: 31, 166]
    print("מחשב חשיבות משקולות (מטריצת פישר) למשימה א'...")
    ewc_module = EWC(model, device)
    ewc_module.compute_fisher(replay_buffer_a)

    # שלב 3: אימון על המשימה השנייה (למשל המשחק Space Invaders)
    print("מתחיל אימון על משימה ב' עם הגנת EWC...")
    # הגדרת ה-lambda (מקדם החשיבות של הזיכרון) [cite: 70, 373]
    importance = 400

    # רשימות לשמירת נתוני הגרף
    epochs = []
    current_losses = []
    ewc_penalties = []

    for epoch in range(1000):
        # אימון על דאטה של משימה ב'
        # נניח דגימת באץ' מבאפר ב':
        # state, action, reward, next_state, done = replay_buffer_b.sample(32)

        optimizer.zero_grad()

        # חישוב ה-Loss של המשימה הנוכחית
        # loss_current = compute_dqn_loss(...)
        loss_current = torch.tensor(1.0, requires_grad=True) # דוגמה

        # הוספת ה"קפיץ" האלסטי שמונע מהמשקולות החשובות של משימה א' להשתנות [cite: 68, 173]
        ewc_penalty = importance * ewc_module.penalty(model)

        total_loss = loss_current + ewc_penalty
        total_loss.backward()
        optimizer.step()

        # שמירת נתונים לגרף
        epochs.append(epoch)
        current_losses.append(loss_current.item())
        ewc_penalties.append(ewc_penalty.item())

        if epoch % 100 == 0:
            print(f"Epoch {epoch}: Current Loss: {loss_current.item():.4f}, EWC Penalty: {ewc_penalty.item():.4f}")

    print("האימון הסתיים. המודל שמר על ביצועי משימה א' תוך למידת משימה ב'.")

    # --- בניית גרפים ---
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1) # גרף ל-Current Loss
    plt.plot(epochs, current_losses, label='Current Loss')
    plt.title('Current Loss Over Epochs')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.grid(True)
    plt.legend()

    plt.subplot(1, 2, 2) # גרף ל-EWC Penalty
    plt.plot(epochs, ewc_penalties, label='EWC Penalty', color='orange')
    plt.title('EWC Penalty Over Epochs')
    plt.xlabel('Epoch')
    plt.ylabel('Penalty')
    plt.grid(True)
    plt.legend()

    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    train_continual_agent()
```

### Atari Environment Preprocessing

To effectively train an agent on Atari environments, observations often require preprocessing. This typically includes:

*   **NoopResetEnv**: Performs a random number of no-op actions after reset to introduce stochasticity.
*   **MaxAndSkipEnv**: Skips frames and takes the element-wise maximum over the last two frames to handle flickering.
*   **EpisodicLifeEnv**: Treats each loss of a life as the end of an episode.
*   **FireResetEnv**: Presses the 'fire' button to start games that require it.
*   **WarpFrame**: Converts frames to grayscale and resizes them to 84x84 pixels.
*   **ScaledFloatFrame**: Scales pixel values to [0, 1].
*   **FrameStack**: Stacks multiple consecutive frames together to capture motion.

The following code defines these wrappers and a `make_atari_env` function to easily create a preprocessed Atari environment.

```python
import gymnasium as gym
from gymnasium import spaces
import numpy as np
from collections import deque
import cv2

class NoopResetEnv(gym.Wrapper):
    def __init__(self, env, noop_max=30):
        super().__init__(env)
        self.noop_max = noop_max
        self.noop_action = 0

    def reset(self, **kwargs):
        self.env.reset(**kwargs)
        noops = self.unwrapped.np_random.integers(1, self.noop_max + 1)
        obs = None
        for _ in range(noops):
            obs, _, terminated, truncated, _ = self.env.step(self.noop_action)
            if terminated or truncated:
                obs, _ = self.env.reset(**kwargs)
        return obs, {}

class MaxAndSkipEnv(gym.Wrapper):
    def __init__(self, env, skip=4):
        super().__init__(env)
        self._obs_buffer = deque(maxlen=2)
        self._skip = skip

    def step(self, action):
        total_reward = 0.0
        terminated = False
        truncated = False
        info = {}
        for _ in range(self._skip):
            obs, reward, terminated, truncated, info = self.env.step(action)
            self._obs_buffer.append(obs)
            total_reward += reward
            if terminated or truncated:
                break
        max_frame = np.max(np.stack(self._obs_buffer), axis=0)
        return max_frame, total_reward, terminated, truncated, info

    def reset(self, **kwargs):
        self._obs_buffer.clear()
        obs, info = self.env.reset(**kwargs)
        self._obs_buffer.append(obs)
        return obs, info

class WarpFrame(gym.ObservationWrapper):
    def __init__(self, env):
        super().__init__(env)
        self.width = 84
        self.height = 84
        self.observation_space = spaces.Box(low=0, high=255, shape=(self.height, self.width, 1), dtype=np.uint8)

    def observation(self, frame):
        frame = cv2.cvtColor(frame, cv2.COLOR_RGB2GRAY)
        frame = cv2.resize(frame, (self.width, self.height), interpolation=cv2.INTER_AREA)
        return frame[:, :, None]

class ScaledFloatFrame(gym.ObservationWrapper):
    def observation(self, obs):
        return np.array(obs).astype(np.float32) / 255.0

class FrameStack(gym.Wrapper):
    def __init__(self, env, k):
        super().__init__(env)
        self.k = k
        self.frames = deque([], maxlen=k)
        shp = env.observation_space.shape
        self.observation_space = spaces.Box(low=0, high=1.0, shape=(shp[0], shp[1], shp[2] * k), dtype=np.float32)

    def reset(self, **kwargs):
        obs, info = self.env.reset(**kwargs)
        for _ in range(self.k):
            self.frames.append(obs)
        return self._get_ob(), info

    def step(self, action):
        obs, reward, terminated, truncated, info = self.env.step(action)
        self.frames.append(obs)
        return self._get_ob(), reward, terminated, truncated, info

    def _get_ob(self):
        return np.concatenate(list(self.frames), axis=-1)

def make_atari_env(env_id='BreakoutNoFrameskip-v4'):
    env = gym.make(env_id, render_mode=None)
    env = NoopResetEnv(env, noop_max=30)
    env = MaxAndSkipEnv(env, skip=4)
    env = WarpFrame(env)
    env = ScaledFloatFrame(env)
    env = FrameStack(env, 4)
    return env
```
